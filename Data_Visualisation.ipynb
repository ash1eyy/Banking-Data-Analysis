# Example of Data Visualisation

This illustration shows the demographics of term deposit subscribers. The graph below highlights the age groups with the highest and lowest subscription rates.

```python


# Display the first few rows to ensure it is loaded correctly
print(df.head())

# Create a histogram for the 'age' variable
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=30, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

By analyzing this graph, we can conclude that candidates between the ages of 30 and 40 have the highest subscription rate. Additionally, the graph exhibits a right skew. This means the data is clustered towards the left side of the graph, with a longer tail extending towards the right side.

Another valuable data visualization tool is the scatterplot graph. Unlike the histogram we saw earlier, scatterplots excel at comparing two variables simultaneously.

```python
# Create a scatterplot for 'balance' vs 'age'
plt.figure(figsize=(12, 6))
sns.scatterplot(x='age', y='balance', data=df)
plt.title('Balance vs Age')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.show()
```

```python
# Create a scatterplot for 'balance' vs 'job'
age_bins = [0, 20, 30, 40, 50, 60, 70, 100]
age_labels = ['<20', '20-29', '30-39', '40-49', '50-59', '60-69', '70+']

# Create a new column 'age_group'
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Create a bar chart for 'Target' vs 'age_group'
plt.figure(figsize=(16, 8))
sns.countplot(x='Target', hue='age_group', data=df)
plt.title('Target Distribution by Age Group')
plt.xlabel('Target')
plt.ylabel('Count')
plt.legend(title='Age Group')
plt.show()
```

We can see that the boxplot has a weak positive correlation as the balance grows when age increases.

```python
# Create a boxplot for the 'balance' variable
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['balance'])
plt.title('Boxplot of Customer Balances')
plt.xlabel('Balance')
plt.show()
```

```python

# Univariate Analysis: Distribution of each variable
# Histogram for numerical columns
df.hist(bins=30, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Columns')
plt.show()


```

```python
# Count plots for categorical columns
categorical_columns = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'Target']
for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=column, data=df, order=df[column].value_counts().index)
    plt.title(f'Count Plot of {column}')
    plt.show()
```

```python
# Boxplot for balance across different jobs
plt.figure(figsize=(12, 8))
sns.boxplot(x='job', y='balance', data=df)
plt.xticks(rotation=45)
plt.title('Balance Distribution Across Different Jobs')
plt.show()
```

```python

# Pairplot to see pairwise relationships
sns.pairplot(df[['age', 'balance', 'duration', 'campaign', 'previous', 'pdays']])
plt.show()
```

```python

# Correlation Matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
